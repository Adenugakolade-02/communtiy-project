#### For some feature engineering

```python
prev['closeddate'] = pd.to_datetime(prev['closeddate'])
prev['firstduedate'] = pd.to_datetime(prev['firstduedate'])
prev['firstrepaiddate'] = pd.to_datetime(prev['firstrepaiddate'])
```

```python
pf_data.loc[pf_data['referredby'].notnull(), 'referredby'] = 1
pf_data.loc[pf_data['referredby'].isnull(), 'referredby'] = 0
```

```python
train = pd.merge(pf_data,prev,on='customerid',how='inner')
```

```python
train.head()
```

```python
train.loc[train['good_bad_flag']=='Good','good_bad_flag'] = 1
train.loc[train['good_bad_flag']=='Bad','good_bad_flag'] = 0
```

```python
train['approveddate'] = pd.to_datetime(train['approveddate'])
train['creationdate'] = pd.to_datetime(train['creationdate'])
train['customerid'] = train['customerid'].astype('category')
```

```python
train['approveddate'] = [datetime.timestamp(i) for i in train['approveddate']]
train['creationdate'] = [datetime.timestamp(i) for i in train['creationdate']]
```

```python
train = train.groupby(train['customerid']).mean().reset_index()
```

```python
target = train['good_bad_flag']
train = train.drop('customerid',axis=1)
del(train['good_bad_flag'])
```

### Using Keras

```python
#Using sigmoid function for my output since we are dealing with binary classification
```

```python
#importing tensorflow and Keras
import tensorflow as tf

import tensorflow.compat.v1 as tf
from tensorflow.keras.backend import clear_session
clear_session()
from tensorflow.compat.v1 import keras
```

```python
from tensorflow.keras import layers
import datetime
logdir = "logs/scalars/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)

#Building the neural network using layers of 400 and 200
model = tf.keras.Sequential()
model.add(layers.Dense(400,input_shape=(11,), activation='relu'))
model.add(layers.Dense(200, activation='relu'))

#Using sigmoid function for my output since we are dealing with binary classification
model.add(layers.Dense(2, activation = 'sigmoid'))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate = 0.01),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
from tensorflow.keras.utils import to_categorical
y_binary = to_categorical(y_train)
y_k = to_categorical(y_test)
model.fit(X_train, y_binary, epochs=50, batch_size=50,
              validation_data=(X_test, y_k),
              callbacks=[tensorboard_callback])
```

```python
print('the NeuralNetwork 1-accuracy score of train set is: {}'.format(1- (accuracy_score(y_train, np.argmax(model.predict(X_train),axis=1)))))
print('the NeuralNetwork 1-accuracy score of test set is: {}'.format(1- (accuracy_score(y_test, np.argmax(model.predict(X_test),axis=1)))))
```
